```python
import json, os

from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_KEY')
os.environ['SERPER_API_KEY'] = userdata.get('SERP_API_KEY')
```

```python
from crewai import Agent, Task, Crew, Process
from langchain_openai.chat_models import ChatOpenAI
from crewai_tools import SerperDevTool
```

```python
llm = ChatOpenAI(model_name="gpt-4o", api_key=os.environ["OPENAI_API_KEY"])
```

```python
search_tool= SerperDevTool()
```

```python
research_agent = Agent(
    llm=llm,
    role='Senior AI Healthcare Researcher',
    backstory=(
        "With a deep understanding of both artificial intelligence and the healthcare industry, "
        "you have spent years exploring how cutting-edge technologies can revolutionize patient care. "
        "Your work bridges the gap between tech innovation and medical application, "
        "making you a pioneer in AI-driven healthcare solutions. "
        "You are driven by a passion for improving patient outcomes and enhancing medical practices through AI."
    ),
    goal=(
        "Investigate the latest developments in AI-powered healthcare innovations, "
        "focusing on technologies that are poised to transform patient care, "
        "from predictive analytics to personalized treatment plans. "
        "Your objective is to compile a comprehensive report that outlines the most promising innovations "
        "and their potential impact on the healthcare industry."
    ),
    tools=[search_tool],
    verbose=True,
    allow_delegation=False
)

research_task = Task(
    description=(
        "Conduct in-depth research on the latest AI-powered innovations in healthcare. "
        "Identify key technologies and their applications, focusing on predictive analytics, personalized medicine, "
        "robotic surgeries, and virtual health assistants. Analyze current trends, emerging startups, and major players "
        "in the AI healthcare space. Provide detailed findings that include data points, insights, and potential impacts "
        "on the healthcare industry."
    ),
    expected_output=(
        "A structured research document that outlines the key findings, including summaries of each AI-driven innovation, "
        "their applications, benefits, challenges, and real-world examples."
    ),
    agent=research_agent
)
```

```python
writer_agent = Agent(
    llm=llm,
    role='Technology Writer',
    backstory=(
        "You are a seasoned writer with a knack for turning complex technical information into engaging and accessible content. "
        "With experience in journalism and content creation, your writing helps bridge the gap between technology experts and the general public. "
        "Your work is characterized by clarity, compelling narratives, and a deep understanding of the subject matter."
    ),
    goal=(
        "Craft a well-written and insightful research article based on research findings, "
        "with a focus on AI-powered innovations in healthcare. "
        "Your article should be engaging, informative, and easy to understand, targeting both industry professionals and the general audience."
    ),
    tools=[search_tool],
    verbose=True,
    allow_delegation=False
)


writing_task = Task(
    description=(
        "Using the research findings provided by the researcher agent, write a compelling and informative article on AI-powered healthcare innovations. "
        "Focus on creating an engaging narrative that highlights the significance of these technologies and their potential impact on the industry. "
        "Ensure that the article is accessible to a broad audience."
    ),
    expected_output=(
        "A deatiled article that discusses AI-driven healthcare innovations, formatted for publication. "
        "The article should include an introduction, main body with detailed sections, and a conclusion that ties together the key points. "
        "The content should be well-organized, clear, and engaging, ready for publication or further editing."
    ),
    agent=writer_agent,
    output_file='result.txt'
)
```

```python
crew = Crew(
    agents=[research_agent, writer_agent],
    tasks=[research_task, writing_task],
    process=Process.sequential
)

result = crew.kickoff()
```
